Here I will be using k - means clustering. it is also called unsupervised algorithm. k - means clustering is the most suitable machine learning for this dataset.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

Before applying the cluster we need to identify the number of cluster. Here i will be using  Silhouette Score method.

```python
from sklearn.metrics import silhouette_score
```

here we will be checking the null values in the dataset

```python
fashion_subset1.isna().sum()
```

Now we will extract the 2 column for clustering I will be using price and avg_rating

```python
fashion_subset1_new = fashion_subset1 [['price','avg_rating']].values
```

Determing the number of clusters

```python
limit = int((fashion_subset1_new.shape[0]//2)**0.5)

# determining number of clusters
# using silhouette score method
for k in range(2, limit+1):
    model = KMeans(n_clusters=k)
    model.fit(fashion_subset1_new)
    pred = model.predict(fashion_subset1_new)
    score = silhouette_score(fashion_subset1_new, pred)
    print('Silhouette Score for k = {}: {:<.3f}'.format(k, score))
```

Now we have the clustering value , so we can start with our k means clustering

```python
# clustering the data using Kmeans
# using k = 2
model = KMeans(n_clusters=2)
 
# predicting the clusters
pred = model.fit_predict(fashion_subset1_new)
 
# plotting all the clusters
colours = ['pink', 'brown']
 
for i in np.unique(model.labels_):
    plt.scatter(fashion_subset1_new[pred==i, 0],
                fashion_subset1_new[pred==i, 1],
                c = colours[i])
     
# plotting the cluster centroids
plt.scatter(model.cluster_centers_[:, 0],
            model.cluster_centers_[:, 1],
            s = 250,  # marker size
            c = 'yellow')
 
plt.title('K Means clustering')
plt.xlabel('Price')
plt.ylabel('Average ratings')
plt.show()
```

Calculate Sum of Square Error(SSE) and elbow method


```python
k_range=range(1,10)
sse=[]
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(fashion_subset1[['price', 'avg_rating']])
    sse.append(km.inertia_) 

sse

plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(k_range,sse)
```
